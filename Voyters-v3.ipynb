```python
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import matplotlib.patches as mpatches

# Set up the canvas - Slightly taller to accommodate spacing
fig, ax = plt.subplots(figsize=(15, 20))
ax.set_xlim(0, 15)
ax.set_ylim(0, 19)
ax.axis('off')

# Title
ax.text(7.5, 18.0, "The Pan-Voyters Alphabet (Corrected)", fontsize=22, fontweight='bold', ha='center')
ax.text(7.5, 17.5, "Green: Roots | Orange: Modifiers (Enlarged) | Black: Structural", fontsize=12, style='italic', ha='center')

# --- Helper Function ---
def draw_symbol(verts, codes, x_offset, y_offset, color, label, desc, is_modifier=False, scale=1.0):
    # Scale vertices
    scaled_verts = [(v[0] * scale, v[1] * scale) for v in verts]
    # Shift vertices
    shifted_verts = [(v[0] + x_offset, v[1] + y_offset) for v in scaled_verts]
    
    path = mpath.Path(shifted_verts, codes)
    
    # Line width
    lw = 3.0 
    patch = mpatches.PathPatch(path, facecolor='none', edgecolor=color, lw=lw, capstyle='round')
    ax.add_patch(patch)
    
    # Label placement
    ax.text(x_offset + (0.5 * scale), y_offset - 0.6, label, fontsize=11, fontweight='bold', ha='center', color=color)
    ax.text(x_offset + (0.5 * scale), y_offset - 1.1, desc, fontsize=8, ha='center', style='italic', wrap=True, color='black')

# ==========================================
# ROW 1: THE ANCIENT ROOTS (GREEN)
# Y = 15.0
# ==========================================
y_row1 = 15.0

# 1. A (Alif)
verts_a = [(0.5, 0.5), (0.5, 2.0), (0.2, 1.5), (0.8, 1.5)]
codes_a = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.MOVETO, mpath.Path.LINETO]
draw_symbol(verts_a, codes_a, 1.0, y_row1, 'green', "1. A (Alif)", "Vertical Starter")

# 2. B (Bet)
verts_b = [(0.1, 0.5), (0.1, 1.5), (0.9, 1.5), (0.9, 0.5), (0.1, 0.5)]
codes_b = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.CLOSEPOLY]
draw_symbol(verts_b, codes_b, 3.5, y_row1, 'green', "2. B (Bet)", "The House")

# 3. G (Gimel)
verts_g = [(0.2, 0.5), (0.5, 1.8), (0.9, 1.2)]
codes_g = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_g, codes_g, 6.0, y_row1, 'green', "3. G (Gimel)", "Camel Angle")

# 4. D (Dalet)
verts_d = [(0.1, 0.5), (0.5, 1.5), (0.9, 0.5), (0.1, 0.5)]
codes_d = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.CLOSEPOLY]
draw_symbol(verts_d, codes_d, 8.5, y_row1, 'green', "4. D (Dalet)", "The Door")

# 5. K (Kaf) - Fixed: Using Lines instead of broken curves
verts_k = [(0.1, 1.5), (0.5, 0.5), (0.9, 1.5)]
codes_k = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_k, codes_k, 11.0, y_row1, 'green', "5. K (Kaf)", "Open Palm")

# ==========================================
# ROW 2: ROOTS CONTINUED (GREEN)
# Y = 12.0
# ==========================================
y_row2 = 12.0

# 6. L (Lam) - Fixed: Removed faulty curve command, used sharp lines for Crook
verts_l = [(0.8, 2.0), (0.8, 0.5), (0.2, 0.5)] 
codes_l = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_l, codes_l, 1.0, y_row2, 'green', "6. L (Lam)", "Crook")

# 7. M (Mem)
verts_m = [(0, 0.5), (0.2, 1.0), (0.4, 0.5), (0.6, 1.0), (0.8, 0.5), (1.0, 0.5)]
codes_m = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_m, codes_m, 3.5, y_row2, 'green', "7. M (Mem)", "Water")

# 8. N (Nun)
verts_n = [(0.2, 1.0), (0.5, 0.5), (0.8, 1.5)]
codes_n = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_n, codes_n, 6.0, y_row2, 'green', "8. N (Nun)", "Fish/Snake")

# 9. R (Resh) - Fixed: Using lines for clearer shape
verts_r = [(0.2, 0.5), (0.2, 1.5), (0.8, 1.5), (0.8, 1.0)]
codes_r = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_r, codes_r, 8.5, y_row2, 'green', "9. R (Resh)", "Head")

# 10. O (Eye) - Fixed: Using Ellipse Patch instead of path
ellipse = mpatches.Ellipse((11.5, y_row2 + 1.0), width=0.8, height=0.8, edgecolor='green', facecolor='none', lw=3.0)
ax.add_patch(ellipse)
ax.text(11.5, y_row2 - 0.6, "10. O (Eye)", fontsize=11, fontweight='bold', ha='center', color='green')
ax.text(11.5, y_row2 - 1.1, "Eye", fontsize=8, ha='center', style='italic', color='black')

# ==========================================
# ROW 3: MODIFIERS (ORANGE) - MOVED DOWN
# Y = 7.5 (Was 9.0)
# ==========================================
y_row3 = 7.5
mod_scale = 2.5 

# 11. Vowel 'i' (Dash Below)
ax.plot([1.0, 3.0], [y_row3+1.0, y_row3+1.0], color='lightgrey', linestyle='--', lw=2)
verts_i = [(0.2, 0.2), (0.6, 0.0)]
codes_i = [mpath.Path.MOVETO, mpath.Path.LINETO]
draw_symbol(verts_i, codes_i, 1.0, y_row3, 'orange', "11. 'i' (Dash)", "Below Line", is_modifier=True, scale=mod_scale)

# 12. Vowel 'u' (Loop Above)
ax.plot([3.5, 5.5], [y_row3+0.5, y_row3+0.5], color='lightgrey', linestyle='--', lw=2)
# Fixed curve geometry
verts_u = [(0.3, 0.2), (0.1, 0.4), (0.3, 0.6), (0.5, 0.4), (0.3, 0.2)] 
codes_u = [mpath.Path.MOVETO, mpath.Path.CURVE3, mpath.Path.CURVE3, mpath.Path.CURVE3, mpath.Path.CURVE3]
# Using simplified diamond shape for stability if curve fails, but curve3 logic fixed here
verts_u_simple = [(0.3, 0.2), (0.1, 0.4), (0.3, 0.6), (0.5, 0.4), (0.3, 0.2)]
codes_u_simple = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_u_simple, codes_u_simple, 3.5, y_row3+0.8, 'orange', "12. 'u' (Loop)", "Above Line", is_modifier=True, scale=mod_scale)

# 13. Hardener (W shape)
ax.plot([6.0, 8.0], [y_row3+0.5, y_row3+0.5], color='lightgrey', linestyle='--', lw=2)
verts_hard = [(0.1, 0.4), (0.2, 0.2), (0.3, 0.4), (0.4, 0.2), (0.5, 0.4)]
codes_hard = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_hard, codes_hard, 6.0, y_row3+0.8, 'orange', "13. Hardener", "Doubles Sound", is_modifier=True, scale=mod_scale)

# 14. Softener (Cup)
ax.plot([8.5, 10.5], [y_row3+0.5, y_row3+0.5], color='lightgrey', linestyle='--', lw=2)
# Fixed: Using lines for 'v' shape cup
verts_soft = [(0.1, 0.4), (0.3, 0.2), (0.5, 0.4)]
codes_soft = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_soft, codes_soft, 8.5, y_row3+0.8, 'orange', "14. Softener", "Softens Sound", is_modifier=True, scale=mod_scale)

# 15. Stop (Circle)
ax.plot([11.0, 13.0], [y_row3+0.5, y_row3+0.5], color='lightgrey', linestyle='--', lw=2)
circle = mpatches.Circle((12.0, y_row3+1.5), 0.3, color='none', ec='orange', lw=3)
ax.add_patch(circle)
ax.text(12.0, y_row3 - 0.6, "15. Stop", fontsize=11, fontweight='bold', ha='center', color='orange')
ax.text(12.0, y_row3 - 1.1, "Full Stop", fontsize=8, ha='center', style='italic', color='black')

# ==========================================
# ROW 4: STRUCTURAL (BLACK) - MOVED DOWN
# Y = 4.0 (Was 5.5)
# ==========================================
y_row4 = 4.0

# 16. Sh (Trident)
verts_s = [(0.1, 1.0), (0.1, 0.5), (0.3, 0.5), (0.3, 1.0), (0.3, 0.5), (0.5, 0.5), (0.5, 1.0), (0.5, 0.5), (0.7, 0.5), (0.7, 1.0)]
codes_s = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_s, codes_s, 1.0, y_row4, 'black', "16. Sh (Trident)", "Teeth")

# 17. T (Cross)
verts_t = [(0.2, 1.5), (0.8, 1.5), (0.5, 1.5), (0.5, 0.5)]
codes_t = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.MOVETO, mpath.Path.LINETO]
draw_symbol(verts_t, codes_t, 3.5, y_row4, 'black', "17. T (Tav)", "Cross")

# 18. P (Gate)
verts_p = [(0.2, 1.5), (0.8, 1.5), (0.8, 0.5), (0.2, 0.5)]
codes_p = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_p, codes_p, 6.0, y_row4, 'black', "18. P (Pe)", "Gate")

# 19. H (Fence)
verts_h = [(0.2, 0.5), (0.2, 1.5), (0.8, 1.5), (0.8, 0.5), (0.2, 1.0), (0.8, 1.0)]
codes_h = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.MOVETO, mpath.Path.LINETO]
draw_symbol(verts_h, codes_h, 8.5, y_row4, 'black', "19. H (Fence)", "Ladder")

# 20. Q (Deep K) - Fixed: Using lines for loop and tail
verts_q = [(0.5, 1.5), (0.8, 1.2), (0.5, 0.9), (0.2, 1.2), (0.5, 1.5), (0.5, 0.0)]
codes_q = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_q, codes_q, 11.0, y_row4, 'black', "20. Q (Qof)", "Tail Down")

# ==========================================
# ROW 5: UNIQUE (BLACK) - MOVED DOWN
# Y = 1.0 (Was 2.0)
# ==========================================
y_row5 = 1.0

# 21. Th (Wheel)
circle = mpatches.Circle((1.5, y_row5+1.0), 0.4, color='none', ec='black', lw=3.0)
ax.add_patch(circle)
ax.plot([1.5, 1.5], [y_row5+0.6, y_row5+1.4], color='black', lw=3.0)
ax.text(1.5, y_row5 - 0.4, "21. Th (Wheel)", fontsize=11, fontweight='bold', ha='center', color='black')
ax.text(1.5, y_row5 - 0.9, "Theta", fontsize=8, ha='center', style='italic', color='black')

# 22. Z (Weapon)
verts_z = [(0.3, 1.5), (0.7, 1.5), (0.5, 1.5), (0.5, 0.5), (0.3, 0.5)]
codes_z = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_z, codes_z, 4.0, y_row5, 'black', "22. Z (Zayin)", "I-Beam")

# 23. Gh (Twist) - Fixed: Using crossed lines
verts_gh = [(0.2, 1.5), (0.8, 0.5), (0.8, 1.5), (0.2, 0.5)]
codes_gh = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.MOVETO, mpath.Path.LINETO]
draw_symbol(verts_gh, codes_gh, 6.5, y_row5, 'black', "23. Gh (Twist)", "X-Shape")

# 24. Ts (Hook)
verts_ts = [(0.2, 1.5), (0.2, 0.5), (0.8, 0.5), (0.8, 1.0)]
codes_ts = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_ts, codes_ts, 9.0, y_row5, 'black', "24. Ts (Tsade)", "Plant Hook")

# 25. Y (Hand)
verts_j = [(0.8, 1.5), (0.8, 0.5), (0.2, 0.5), (0.2, 1.0)]
codes_j = [mpath.Path.MOVETO, mpath.Path.LINETO, mpath.Path.LINETO, mpath.Path.LINETO]
draw_symbol(verts_j, codes_j, 11.5, y_row5, 'black', "25. Y (Hand)", "Rev Hook")

plt.tight_layout()
plt.savefig('full_pan_voyters_alphabet_v3.png')
print("High-Visibility Alphabet Generated (Corrected Layout & Symbols).")
plt.show()
```
